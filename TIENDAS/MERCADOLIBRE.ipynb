```python
# Importaciones
import time 
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Inicializar la lista de datos
datas = []

# Inicializar el driver de Selenium
driver = webdriver.Chrome()

# Iterar a través del rango
for i in range(0, 5):

    # Construir la URL
    if i == 0:
        url = 'https://listado.mercadolibre.com.pe/celulares-telefonos/celulares-smartphones/apple/celulares_NoIndex_True'
    else:
        url = f'https://listado.mercadolibre.com.pe/celulares-telefonos/celulares-smartphones/apple/celulares_Desde_{i*50+1}_NoIndex_True'
    
    # Navegar a la URL
    driver.get(url)

    # Encontrar la tabla en la página
    table = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, '//*[@id="root-app"]/div/div[2]/section/ol')))
    table_rows = table.find_elements(By.TAG_NAME, 'li')
    print(f"Found {len(table_rows)} rows on the table from page {i}")

    tablex = []

    # Iterar a través de las filas de la tabla
    for row in table_rows:
        # Usar los selectores CSS correctos para su sitio web
        title_elements = row.find_elements(By.CSS_SELECTOR, ".ui-search-item__title")
        price_elements = row.find_elements(By.CSS_SELECTOR, ".andes-money-amount__fraction")
        rating_elements = row.find_elements(By.CSS_SELECTOR, ".ui-search-reviews__amount")
        available_colors_elements = row.find_elements(By.CSS_SELECTOR, ".ui-search-item__group__element.ui-search-item__variations-text.shops__items-group-details")
        link_elements = row.find_elements(By.CSS_SELECTOR, ".ui-search-link")

        # Extraer el texto de los elementos encontrados
        title = title_elements[0].text if title_elements else None
        price = price_elements[0].text if price_elements else None
        rating = rating_elements[0].text if rating_elements else None
        available_colors = available_colors_elements[0].text if available_colors_elements else None
        link = link_elements[0].get_attribute('href') if link_elements else None

        # Agregar los datos a la lista
        tablex.append({"Title": title, "Price": price, "Rating": rating, "Available_colors": available_colors,"Link":link})

    # Convertir la lista en un DataFrame de pandas y agregarlo a la lista de datas
    df = pd.DataFrame(tablex)
    datas.append(df)

# Cerrar el driver de Selenium
driver.quit()
```

```python
final_df = pd.concat(datas, ignore_index=True)
final_df
```

```python
final_df.to_csv("hola.csv")
```
