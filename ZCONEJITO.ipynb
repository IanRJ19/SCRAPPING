```python
# Importaciones
#paquetes y librerias
import time 
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options


# Inicializar la lista de datos
datas = []

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")  # Ajusta las dimensiones según lo necesites

# Inicializar el driver de Selenium con las opciones definidas
driver = webdriver.Chrome(options=chrome_options)
# Iterar a través del rango
for i in range(0, 1):

    # Construir la URL
    if i == 0:
        url = 'https://jp.mercari.com/search?keyword=bts'
    else:
        url = f'https://jp.mercari.com/search?keyword=bts&page_token=v1%3A{i}'
    
    # Navegar a la URL
    driver.get(url)

    # Encontrar la tabla en la página
    test=WebDriverWait(driver, 100).until(EC.presence_of_element_located((By.CLASS_NAME, "itemName__a6f874a2")))
    table = WebDriverWait(driver, 100).until(EC.presence_of_element_located((By.XPATH, '//*[@id="item-grid"]')))
    table_rows = table.find_elements(By.TAG_NAME, 'li') 
    print(f"Found {len(table_rows)} rows on the table from page {i}")

    tablex = []
    # Iterar sobre cada fila y obtener el texto
    for index, row in enumerate(table_rows):

        print(f"Row {index}: {row.text}")

    # Iterar a través de las filas de la tabla
    for row in table_rows:
        # Usar los selectores CSS correctos para su sitio web
        title_elements = row.find_elements(By.CLASS_NAME, "itemName__a6f874a2") 
        price_elements = row.find_elements(By.CLASS_NAME, "number__6b270ca7")
        image_elements = row.find_elements(By.TAG_NAME, "img")

        link_elements = row.find_elements(By.CSS_SELECTOR, 'a[data-location="search_result:best_match:body:item_list:item_thumbnail"]')

        # Extraer el texto de los elementos encontrados
        title = title_elements[0].text if title_elements else None
        price = price_elements[0].text if price_elements else None
        image = image_elements[0].get_attribute('src') if image_elements else None
        if image:
            image = f'<img src="{image}" width="100" >'
        link = link_elements[0].get_attribute('href') if link_elements else None
        # Imprimir los detalles de cada fila
        print(f'Title: {title}, Price: {price}, Link: {link}')

        # Agregar los datos a la lista
        tablex.append({"Title": title, "Price": price, "Link": link, "Imagen": image})

    # Convertir la lista en un DataFrame de pandas y agregarlo a la lista de datas
    df = pd.DataFrame(tablex)
    datas.append(df)

# Cerrar el driver de Selenium
#driver.quit()
```

```python
final_df = pd.concat(datas, ignore_index=True)
final_df
```

```python
import requests

# Obtén la tasa de cambio más reciente de JPY a PEN usando una API de terceros
api_url = "https://open.er-api.com/v6/latest/JPY"
response = requests.get(api_url)
data = response.json()
tasa_cambio = data['rates']['PEN']

# Convierte los precios a números y luego a soles peruanos
final_df['Price'] = final_df['Price'].astype(str).str.replace(',', '').astype(float)
final_df['Price'] = (final_df['Price'] * tasa_cambio).round(2)

# Muestra el DataFrame actualizado
print(final_df)
```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.drawing.image import Image
from PIL import Image as PILImage
import io
import requests
from IPython.display import display, HTML

# Crear un nuevo libro de trabajo de Excel y seleccionar la hoja activa
wb = Workbook()
ws = wb.active

# Agrega los encabezados de columna al archivo Excel
for col_num, col_name in enumerate(final_df.columns):
    col_letter = chr(col_num + 65)
    ws[f"{col_letter}1"] = col_name

# Agrega los datos y las imágenes al archivo Excel
for row_num, row_data in final_df.iterrows():
    for col_num, col_data in enumerate(row_data):
        col_letter = chr(col_num + 65)
        
        # Insertar datos de texto
        if col_num != 3:  # Ignorando la columna de imágenes por ahora
            ws[f"{col_letter}{row_num + 2}"] = col_data
        
        else:  # Insertando imágenes
            img_url = col_data.split('"')[1]
            response = requests.get(img_url)
            img = PILImage.open(io.BytesIO(response.content))

            # Cambia el tamaño de la imagen antes de guardarla en un objeto BytesIO
            img_width, img_height = img.size
            scale_factor = 0.5  # Ajusta este valor según sea necesario
            new_width = int(img_width * scale_factor)
            new_height = int(img_height * scale_factor)
            img = img.resize((new_width, new_height), PILImage.BICUBIC)

            # Guarda la imagen en un objeto BytesIO
            img_byte_arr = io.BytesIO()
            img.save(img_byte_arr, format='PNG')
            
            # Crea una imagen openpyxl a partir del objeto BytesIO y agrégala a la hoja de Excel
            img = Image(img_byte_arr)
            ws.add_image(img, f"E{row_num + 2}")

# Guardar el archivo Excel
wb.save("output.xlsx")
```
